# Une suite récurrente

On définit sur $[0,1]$ l'application $f:x\longmapsto \frac{1}{2}x(1-x)$. Le but de ce problème est d'étudier les suites récurrentes définies par la condition initiale $u_0\in[0,1]$ et $u_{n+1}=f(u_n)$ pour tout $n\in\mathbb N$.

```python
%matplotlib notebook
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 0.5*x*(1-x)

U = [0.4]
n = 20
for i in range(n):
    U.append(f(U[-1]))
fig, ax = plt.subplots(2,1)

ax[0].plot(U, 'r.:', label=r'$u_n$')
ax[0].legend()
ax[0].grid()

ax[1].plot([u*2**i for i,u in enumerate(U)], 'b.:', label = r'$2^nu_n$')
ax[1].legend()
ax[1].grid()

fig.tight_layout()
fig.show()
```
