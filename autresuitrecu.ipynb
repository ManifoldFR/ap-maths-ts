```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact,widgets
```

```python
# Définition de f

def plotu(a):
    def F(x):
        return a*np.sin(x)

    U = [1]
    n = 40

    for i in range(n):
        U.append(F(U[-1]))
    
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(1,1,1)
    ax.grid()
    ax.set_ylim(bottom=-3,top=3)
    ax.plot(U, 'r.:')
    ax.set_xlim()

interact(plotu,a=widgets.FloatSlider(value=1,min=-1.5,max=3.,step=0.02))
```
