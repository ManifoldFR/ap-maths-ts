<li> La fonction $f$ est dérivable, sa dérivée est $f'(x)=3(x-1)^2\geq0$, d'où le tableau de signe 
<table>
<tr> <td>$x$ </td>    <td> $-\infty\qquad1$</td>  </tr>
<tr> <td> $f'(x)$ </td> <td> $+$ </td>  </tr>
<tr> <td> $f$ </td>    <td>$\nearrow$ </td> </tr>
</table>
   Et $f(x)\to-\infty$ lorsque $x\to-\infty$, $f(1)=0$. Ainsi l'image de $]-\infty,1]$ par $f$ est $]-\infty,0]\subset{]-\infty,1]}$. D'où la stabilité de $I:={]-\infty,1]}$ par $f$.
   Ainsi $$\forall n\in\mathbb N, \ u_n\in I.$$
</li>
<li> On a $g(x)=(x-1)^3-x$, d'où $g$ est dérivable de dérivée $g'(x)=3(x-1)^2-1 = 3x^2-6x+2 = (x-\frac{\sqrt 3}{3})(x+\frac{\sqrt 3}{3})$. Donc $g$ a donc pour maximum sur $I$ le réel $g(1-\frac{\sqrt 3}{3}) = \frac{2}{3\sqrt 3}-1<0$. Ainsi $g<0$ sur $I$.

   On en déduit que $\forall n\in\mathbb N,\ u_{n+1}-u_n = g(u_n)<0$: $(u_n)$ est strictement décroissante.
</li>
<li> On a deux cas de figure : $u$ minorée (et donc converge vers un réel $\ell < u_0 \leq 1$), et $u$ non minorée (et donc $u_n\to-\infty$).
 
 Si $u$ est minorée, elle admet une limite finie $\ell\in I$ qui vérifie alors $f(\ell)=\ell$ i.e. $g(\ell)=0$, mais $g<0$ sur $I$: absurde.  
 Donc $u$ n'est pas minorée, $$u_n\xrightarrow[n\to+\infty]{}-\infty. $$
</li></ol>

# Exercice I.7

Soit
$$
f:\left\{
\begin{array}{ll}
\mathbb R & \longrightarrow \mathbb R \\
x &\longmapsto -(x-2)^2+2
\end{array}
\right.
$$

Étudions le signe de $g:x\longmapsto f(x)-x$ sur ${]-\infty,2]}$. On a $g(x)=-(x-2)^2+2-x = -x^2+3x-2 = -(x-1)(x-2)$ pour tout $x\< \leq 2$, d'où $1$ et $2$ sont les seuls points fixes de $f$, et le tableau de signe suivant:
<table { 
margin-left: auto;
margin-right: auto;
}>
    <tr>
        <td> x </td> <td> $-\infty\qquad 1$ </td> <td> $\qquad 2$ </td>
    </tr>
    <tr>
        <td> g(x) </td> <td> $-$ </td> <td> $+$ </td>
    </tr>
</table>

De plus $f'(x)=-2(x-2)$ d'où $f$ est strictement croissante sur $]-\infty,2]$. Donc $1<x<2\Rightarrow1=f(1)<f(x)<f(2)=2$ (donc $I:={]-\infty,1[}$ est stable), et $x<1 \Rightarrow f(x)<f(1)=1$ (donc $J:={]1,2[}$ est stable).

On distingue trois cas : $u_0<1$, $u_0=1$, et $1<u_0<2$.

Dans le deuxième cas $u$ stationne à $1$.  
Dans le premier, on a $u_n\in I\; \forall n\in\mathbb N$ donc $u_{n+1}-u_n=g(u_n)<0$ et $u$ est strictement décroissante, non minorée sinon elle converge vers un point fixe qui serait inférieur à 1 ce qui est impossible : donc elle tend vers $-\infty$.  
Dans le dernier, on a $u_n\in J\;\forall n\in\mathbb N$ donc $g(u_n)>0$ et $u$ est strictement croissante, majorée par $2$ ; ainsi $u$ converge vers un point fixe $\ell$ de $f$, qui ne peut être $1$ par croissance de $u$ : $u_n\xrightarrow[n\to+\infty]{}2$.

Pour finir, un petit aperçu du comportement de $(u_n)$ avec diverses valeurs de $u_0$ :

```python
%matplotlib inline
import matplotlib.pyplot as plt

def f(x):
    '''Fonction de récurrence f'''
    return -(x-2)**2+2

def Phi(u0,n):
    '''Construit un graphe des n premiers termes de la suite,
    avec condition initiale u0'''
    fig,ax = plt.subplots(1,1) # définit une instance de graphe
    L = [u0]
    for i in range(n): # construit les n premiers termes
        terme = L[-1]
        term_suiv = f(terme)
        L.append(term_suiv)
    ax.grid()
    ax.margins(y=0.05) # marge additionnelle sur l'axe des ordonnées
    ax.set_xlabel(r'$n$') # titre axe des abscisses
    ax.set_ylabel(r'$u_n$') # titre axe des ordonnées
    ax.plot(L,':.') # ajoute les points de la liste L au graphe
```

```python
Phi(0.9999,15)
```

```python
Phi(1,30)
```

```python
Phi(1.0001,30)
```
